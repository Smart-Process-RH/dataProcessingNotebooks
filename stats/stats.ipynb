```python
applications = requests.get(fetch_application_endpoint, headers={"x-api-key": f"{api_key}"})
print(applications)
print(applications.json())
applications_df = pd.DataFrame(applications.json())

if not os.path.exists('applications'):
    os.makedirs('applications')

applications_df.to_csv('applications/raw_applications.csv', index=False)

cabine_cibli_job_applications = []
for application in applications_df.to_dict(orient='records'):
    if application['source'] == 'cabine cibli job':
        cabine_cibli_job_applications.append(application)

cabine_cibli_job_applications_df = pd.DataFrame(cabine_cibli_job_applications)

cabine_cibli_job_applications_df.to_csv('applications/cabine_cibli_job_applications.csv', index=False)
```

```python
# Count all applications per campaign (from full applications dataset)
all_applications_count = {}
for application in applications_df.to_dict(orient='records'):
    if 'campaign' in application and application['campaign'] is not None and 'id' in application['campaign']:
        campaign_id = application['campaign']['id']
        all_applications_count[campaign_id] = all_applications_count.get(campaign_id, 0) + 1

# Count cibli cabine applications per campaign
cibli_applications_count = {}
for application in cabine_cibli_job_applications:
    if 'campaign' in application and 'id' in application['campaign']:
        campaign_id = application['campaign']['id']
        cibli_applications_count[campaign_id] = cibli_applications_count.get(campaign_id, 0) + 1

# Get all unique campaign IDs from both datasets
all_campaign_ids = set(all_applications_count.keys()) | set(cibli_applications_count.keys())
unique_campaign_ids = sorted(list(all_campaign_ids))

print(f"Total unique campaigns: {len(unique_campaign_ids)}")

# Create a comprehensive DataFrame with both counts
campaign_stats = []
for campaign_id in unique_campaign_ids:
    total_count = all_applications_count.get(campaign_id, 0)
    cibli_count = cibli_applications_count.get(campaign_id, 0)
    campaign_stats.append({
        'campaign_id': campaign_id,
        'total_applications': total_count,
        'cibli_cabine_applications': cibli_count
    })

campaign_stats_df = pd.DataFrame(campaign_stats)
print("\nCampaign Statistics (All Applications vs Cibli Cabine Applications):")
print(campaign_stats_df)

# Display summary statistics
print(f"\nSummary:")
print(f"Total campaigns: {len(campaign_stats_df)}")
print(f"Total applications across all campaigns: {campaign_stats_df['total_applications'].sum()}")
print(f"Total cibli cabine applications across all campaigns: {campaign_stats_df['cibli_cabine_applications'].sum()}")
print(f"Campaigns with cibli applications: {(campaign_stats_df['cibli_cabine_applications'] > 0).sum()}")
```

```python
campaigns = requests.get(fetch_campaigns_endpoint, headers={"x-api-key": f"{api_key}"})
campaigns_data = campaigns.json()
campaigns_df = pd.DataFrame(campaigns_data)

if not os.path.exists('campaigns'):
    os.makedirs('campaigns')

campaigns_df.to_csv('campaigns/raw_campaigns.csv', index=False)

def get_client_identifier(organization):
    if not organization or not isinstance(organization, dict):
        print(f"Failed - organization is invalid: {organization}")
        return 'unknown'
    
    if organization.get('group'):
        group = organization['group']
        return group.get('name') if isinstance(group, dict) else group
    
    if organization.get('brand'):
        brand = organization['brand']
        return brand.get('name') if isinstance(brand, dict) else brand
    
    if organization.get('unit'):
        unit = organization['unit']
        return unit.get('name') if isinstance(unit, dict) else unit
        
    print(f"Failed - no group/brand/unit found: {json.dumps(organization, indent=2, ensure_ascii=False)}")
    return 'unknown'

campaign_to_client = {}
for campaign in campaigns_data:
    campaign_id = campaign.get('id')
    org = campaign.get('organization')
    
    if isinstance(org, str):
        try:
            org = json.loads(org)
        except:
            try:
                org = ast.literal_eval(org)
            except:
                org = None
    
    campaign_to_client[campaign_id] = get_client_identifier(org)

# Debug: Find missing campaign IDs
app_campaign_ids = set()
for app in applications_df.to_dict(orient='records'):
    if app.get('campaign') and app['campaign'].get('id'):
        app_campaign_ids.add(app['campaign']['id'])

fetched_campaign_ids = set(campaign_to_client.keys())
missing_campaign_ids = app_campaign_ids - fetched_campaign_ids

print(f"Fetched campaigns: {len(fetched_campaign_ids)}")
print(f"Campaign IDs in applications: {len(app_campaign_ids)}")
print(f"Missing campaign IDs: {len(missing_campaign_ids)}")
if missing_campaign_ids:
    sample_missing = sorted(list(missing_campaign_ids))[:3]
    print(f"Sample missing IDs: {sample_missing}")
    print("\nFetching sample missing campaigns individually:")
    for missing_id in sample_missing:
        try:
            resp = requests.get(f"{api_url}/campaigns/{missing_id}", headers={"x-api-key": f"{api_key}"})
            if resp.status_code == 200:
                campaign = resp.json()
                org = campaign.get('organization')
                print(f"  Campaign {missing_id}: EXISTS - org type: {type(org)}")
                if org:
                    client_id = get_client_identifier(org if isinstance(org, dict) else json.loads(org) if isinstance(org, str) else None)
                    print(f"    Client identifier: {client_id}")
            else:
                print(f"  Campaign {missing_id}: NOT FOUND (status {resp.status_code})")
        except Exception as e:
            print(f"  Campaign {missing_id}: ERROR - {e}")

all_applications_by_client = {}
for app in applications_df.to_dict(orient='records'):
    if app.get('campaign') and app['campaign'].get('id'):
        campaign_id = app['campaign']['id']
        client = campaign_to_client.get(campaign_id, 'unknown')
        if client == 'unknown' and campaign_id in missing_campaign_ids:
            # This is unknown because campaign wasn't fetched, not because org parsing failed
            pass
        all_applications_by_client[client] = all_applications_by_client.get(client, 0) + 1

cibli_applications_by_client = {}
for app in cabine_cibli_job_applications:
    if app.get('campaign') and app['campaign'].get('id'):
        client = campaign_to_client.get(app['campaign']['id'], 'unknown')
        cibli_applications_by_client[client] = cibli_applications_by_client.get(client, 0) + 1

client_stats_df = pd.DataFrame([
    {
        'client_identifier': client,
        'total_applications': all_applications_by_client.get(client, 0),
        'cibli_cabine_applications': cibli_applications_by_client.get(client, 0)
    }
    for client in sorted(set(all_applications_by_client.keys()) | set(cibli_applications_by_client.keys()))
])

print(client_stats_df)
print(f"\nUnknown applications: {all_applications_by_client.get('unknown', 0)}")




```

```python
# Save client stats to excel (exclure "unknown")
client_stats_filtered = client_stats_df[client_stats_df['client_identifier'] != 'unknown']
client_stats_filtered.to_excel('client_stats.xlsx', index=False)

# Save client stats to csv (exclure "unknown")
client_stats_filtered.to_csv('client_stats.csv', index=False)

print(f"Fichiers sauvegard√©s (excluant 'unknown'): {len(client_stats_filtered)} clients")
```

```python
# Create hierarchical Excel with expandable rows (Client -> Campaigns)
from openpyxl import Workbook
from openpyxl.styles import Font, Alignment, PatternFill
from openpyxl.utils import get_column_letter

# Cr√©er un mapping campaign_id -> campaign_title
campaign_id_to_title = {}
for campaign in campaigns_data:
    campaign_id = campaign.get('id')
    campaign_title = campaign.get('title', f'Campaign {campaign_id}')
    campaign_id_to_title[campaign_id] = campaign_title

# Build hierarchical data structure (exclure "unknown")
hierarchical_data = {}

for app in applications_df.to_dict(orient='records'):
    if app.get('campaign') and app['campaign'].get('id'):
        campaign_id = app['campaign']['id']
        client = campaign_to_client.get(campaign_id, 'unknown')
        
        # Exclure les clients "unknown"
        if client == 'unknown':
            continue
        
        if client not in hierarchical_data:
            hierarchical_data[client] = {}
        
        if campaign_id not in hierarchical_data[client]:
            hierarchical_data[client][campaign_id] = {'total': 0, 'cibli': 0}
        
        hierarchical_data[client][campaign_id]['total'] += 1
        if app.get('source') == 'cabine cibli job':
            hierarchical_data[client][campaign_id]['cibli'] += 1

# Create workbook
wb = Workbook()
ws = wb.active
ws.title = "Client Statistics"

# Headers
headers = ['Client', 'Campaign Title', 'Total Applications', 'Cibli Cabine Applications']
ws.append(headers)

# Style headers
header_fill = PatternFill(start_color="366092", end_color="366092", fill_type="solid")
header_font = Font(bold=True, color="FFFFFF")
for cell in ws[1]:
    cell.fill = header_fill
    cell.font = header_font
    cell.alignment = Alignment(horizontal="center", vertical="center")

# Add data with grouping
row_idx = 2
client_rows = {}  # Track which rows belong to each client group

for client in sorted(hierarchical_data.keys()):
    client_start_row = row_idx
    client_total = 0
    client_cibli = 0
    
    # Add client total row first
    ws.append([client, "TOTAL", 0, 0])  # Will update totals after
    client_row = row_idx
    row_idx += 1
    
    campaign_start_row = row_idx
    
    # Add campaign rows for this client (utiliser le titre au lieu de l'ID)
    for campaign_id in sorted(hierarchical_data[client].keys()):
        data = hierarchical_data[client][campaign_id]
        campaign_title = campaign_id_to_title.get(campaign_id, f'Campaign {campaign_id}')
        ws.append([client, campaign_title, data['total'], data['cibli']])
        client_total += data['total']
        client_cibli += data['cibli']
        row_idx += 1
    
    campaign_end_row = row_idx - 1
    
    # Update client total row
    ws.cell(row=client_row, column=1, value=client).font = Font(bold=True)
    ws.cell(row=client_row, column=2, value="TOTAL").font = Font(bold=True, italic=True)
    ws.cell(row=client_row, column=3, value=client_total).font = Font(bold=True)
    ws.cell(row=client_row, column=4, value=client_cibli).font = Font(bold=True)
    
    # Group the campaign rows under the client (collapsable)
    if campaign_end_row >= campaign_start_row:
        ws.row_dimensions.group(campaign_start_row, campaign_end_row, outline_level=1, hidden=False)
    
    client_rows[client] = (client_row, campaign_start_row, campaign_end_row)

# Adjust column widths
ws.column_dimensions['A'].width = 40
ws.column_dimensions['B'].width = 60  # Plus large pour les titres de campagne
ws.column_dimensions['C'].width = 20
ws.column_dimensions['D'].width = 25

# Freeze header row
ws.freeze_panes = 'A2'

# Enable outline (pour que les groupes soient collapsables par d√©faut)
ws.sheet_properties.outlinePr.summaryBelow = True
ws.sheet_properties.outlinePr.summaryRight = False

# Save
wb.save('client_stats_hierarchical.xlsx')
print("Hierarchical Excel file saved: client_stats_hierarchical.xlsx")
print("You can expand/collapse rows by client in Excel!")
print(f"Total clients (excluding 'unknown'): {len(hierarchical_data)}")
```
